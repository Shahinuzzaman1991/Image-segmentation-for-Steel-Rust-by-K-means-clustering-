```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from skimage import io
```

```python

image = io.imread("/content/drive/MyDrive/3.jpeg")
pixels = image.reshape(-1, 3)
k_values = range(1, 6)
inertia_values = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(pixels)
    inertia_values.append(kmeans.inertia_)
plt.figure(figsize=(6, 5))
plt.plot(k_values, inertia_values, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
# plt.grid(true)
plt.show()
```

```python
optimal_k =2
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
kmeans.fit(pixels)
segmented_image = kmeans.labels_.reshape(image.shape[:2])
plt.figure(figsize=(8, 6))
plt.imshow(segmented_image, cmap='coolwarm')
plt.axis('off')
plt.title(f'Segmented Image (k={optimal_k})')
plt.show()
```

```python
import cv2
output_path = '/content/drive/MyDrive/Colab Notebooks/IMAGE-2.jpg'
segmented_image = segmented_image.astype(np.uint8)
cv2.imwrite(output_path, segmented_image)
print(f'Segmented_image saved as {output_path}')
```
